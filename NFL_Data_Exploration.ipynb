```python
from sklearn.ensemble import RandomForestClassifier

import pandas as pd


passing_plays = play_by_play[play_by_play["play_type_nfl"] == "PASS"].copy().reset_index(drop=True)
numerics = ['int16', 'int32', 'int64', 'float16', 'float32', 'float64', 'boolean']

numeric_columns = passing_plays.select_dtypes(include=numerics)
numeric_columns = numeric_columns.dropna(axis=1)

reward_column = "yards_gained"
reward_set = numeric_columns[reward_column]


data_set = numeric_columns.loc[:, numeric_columns.columns != reward_column]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(data_set, reward_set, test_size=0.33)
```

```python
rf = RandomForestClassifier()
rf.fit(X_train, y_train)
importance = rf.feature_importances_

df = pd.DataFrame(rf.feature_importances_,
                 index = X_train.columns,
                 columns=['importance']).sort_values('importance', ascending=False).reset_index(drop=False)
df.head(20)
```

```python
rf.score(X_test, y_test)
```
